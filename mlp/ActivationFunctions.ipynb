```python
import numpy as np
import matplotlib.pyplot as plt
```

Sigmoid function

```python
x = np.linspace(-5, 5, 1000)
y = 1 / (1 + np.exp(-x))

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/sigmoid.pdf')
```

Tanh function

```python
x = np.linspace(-5, 5, 1000)
y = np.tanh(x)

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/tanh.pdf')
```

ReLU function

```python
x = np.linspace(-5, 5, 1000)
y = np.where(x >= 0, x, 0)

plt.plot(x, y, lw=2)
plt.grid()
plt.savefig('figures/relu.pdf')
```
